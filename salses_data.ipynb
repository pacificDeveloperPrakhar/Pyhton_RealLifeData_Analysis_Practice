```python
import pandas as pd
import numpy as np
```

```python
sales_csv=pd.read_csv('./sales_data.csv')
```

```python
sales_csv
```

```python
sales_csv.groupby("Product Name").describe()
```

```python
sales_csv.groupby("Product Name")["Total Sale"].mean()
```

```python
#grouping by multiple fields 
sales_csv.groupby(["Product Name","Region"]).describe()
```

```python
# aggregate function is used to perform the one or more operations on the data set
sales_csv.agg({"Total Sale":["sum","count"]})
```

```python
sales_csv_new=pd.read_csv('./sales_data_new.csv')
```

```python
pd.concat([sales_csv,sales_csv_new],axis=0)
```

```python
pd.merge(sales_csv,sales_csv_new,on="Region")
```

```python
pd.merge(sales_csv_new.head(),sales_csv.head(),on="Region",how="inner")
```

#### there are four types of joining the data such as inner outer left right
<p>make sure that the data is not too large the above inference took like more than two minutes hence it is very much important to see and make certain that the dataset is not too large</p>

```python
pd.merge(sales_csv_new.head(),sales_csv.head(),on="Region",how="right")
```
